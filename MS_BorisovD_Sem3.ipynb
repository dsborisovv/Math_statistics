```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 3.1

```python
from scipy.special import gamma
x = np.linspace(-4, 6, 2221)
y = gamma(x)
plt.grid(True)
plt.title('Gamma function')
plt.xlim(-4, 6)
plt.ylim(-20, 22)
plt.plot(x, y);
```

```python
from scipy.special import gamma
s = np.linspace(0, 6, 5000)
g = [gamma(i) for i in s]
g_min = min(g)
s_min = [s[i] for i in range(len(s)) if g_min == g[i]]
    
print(g_min, s_min[0])
```

## 3.2

```python
from scipy.stats import norm
rv_norm1 = norm(0, 1)
rv_norm2 = norm(0, 0.7)
rv_norm3 = norm(0, 2)
x = np.linspace(-4, 4, 200)
func1 = rv_norm1.pdf(x)
func2 = rv_norm2.pdf(x)
func3 = rv_norm3.pdf(x)
plt.grid(True)
plt.ylim(0, 0.6)
plt.xlim(-4, 4)
plt.plot(x, func1)
plt.fill_between(x, func1, alpha = 0.15)
plt.plot(x, func2)
plt.fill_between(x, func2, alpha = 0.15)
plt.plot(x, func3)
plt.fill_between(x, func3, alpha = 0.15);
```

```python
from scipy.stats import norm 

rv_norm1 = norm(1, 1)
rv_norm2 = norm(3, 1)
rv_norm3 = norm(-1, 1)
x = np.linspace(-6, 8, 200)
func1 = rv_norm1.pdf(x)
func2 = rv_norm2.pdf(x)
func3 = rv_norm3.pdf(x)
plt.grid(True)
plt.ylim(0, 0.45)
plt.xlim(-6, 8)
plt.plot(x, func1)
plt.fill_between(x, func1, alpha = 0.15)
plt.plot(x, func2)
plt.fill_between(x, func2, alpha = 0.15)
plt.plot(x, func3)
plt.fill_between(x, func3, alpha = 0.15);
```

## 3.3

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 10, 200)
plt.ylim(0, 0.4)
plt.xlim(0, 10)
plt.grid(True)
ax.plot(x, chi2.pdf(x, 2))
ax.fill_between(x, chi2.pdf(x, 2), alpha = 0.15)
ax.plot(x, chi2.pdf(x, 3))
ax.fill_between(x, chi2.pdf(x, 3), alpha = 0.15)
ax.plot(x, chi2.pdf(x, 5))
ax.fill_between(x, chi2.pdf(x, 5), alpha = 0.15)
ax.plot(x, chi2.pdf(x, 10))
ax.fill_between(x, chi2.pdf(x, 10), alpha = 0.15);
```

```python
from scipy.stats import chi2
print(chi2.sf(10.9, df = 20))
print(chi2.cdf(28.9, df = 20))
print(chi2.cdf(31.4, df = 25) - chi2.cdf(8.26, df = 25))
```

```python
from scipy.stats import chi2
print(chi2(5).isf(0.1))
print(chi2(100).isf(0.01))
print(chi2(18).isf(0.95))
print(chi2(80).isf(0.99))
```

## 3.4

```python
from scipy.stats import t

fig, ax = plt.subplots(1, 1)
x = np.linspace(-4, 4, 200)
plt.ylim(0, 0.4)
plt.xlim(-4, 4)
plt.grid(True)
ax.plot(x, t.pdf(x, 1))
ax.fill_between(x, t.pdf(x, 1), alpha = 0.15)
ax.plot(x, t.pdf(x, 2))
ax.fill_between(x, t.pdf(x, 2), alpha = 0.15)
ax.plot(x, t.pdf(x, 4))
ax.fill_between(x, t.pdf(x, 4), alpha = 0.15);

print('P:')
print(t.pdf(2.23, 10))
print(t.cdf(-2.23, 10))
print(t.sf(2.23, 10) - t.sf(-2.23, 10))
print(t.pdf(3.17, 10) - t.pdf(-1.31, 10))
print(t.sf(1.96,df = 82))

print('Percent')
print(t.isf(0.005, 33))
print(t.isf(0.01, 23))
print(t.isf(0.05, 100))
print(t.isf(0.0025, np.inf))
```

## 3.5

```python
from scipy.stats import f

fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 3, 200)
plt.ylim(0, 1)
plt.xlim(0, 3)
plt.grid(True)
ax.plot(x, f.pdf(x, 2, 10))
ax.fill_between(x, f.pdf(x, 2, 10), alpha = 0.15)
ax.plot(x, f.pdf(x, 5, 10))
ax.fill_between(x, f.pdf(x, 5, 10), alpha = 0.15)
ax.plot(x, f.pdf(x, 20, 10))
ax.fill_between(x, f.pdf(x, 20, 10), alpha = 0.15);

print('P:')
print(f.cdf(3.24, 3, 16) - f.cdf(-3.24, 3, 16))
print(f.sf(1.3, 35, 100))

print('Percent')
print(f.isf(0.05, 3, 7))
print(f.isf(0.05, 7, 3))
print(f.isf(0.025, 5, 20))
print(f.isf(0.05, 300, 800))
```
